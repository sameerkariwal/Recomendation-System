#### module that reads both the data sources available for creating recommendations


```python
def reading_content_data():
    
    plat_articles = pd.read_csv(r'C:\Users\SumitJain\Documents\upgd\recommendation_system\data\platform_content.csv')
    plat_articles = plat_articles[plat_articles['interaction_type'] == 'content_present']
#   or alternatively can be done using
#   df_pres = plat_articles.interaction_type.str.contains('content_present', case=False)
#   plat_articles = plat_articles[df_pres]

    user_behaviour_df = pd.read_csv(r'C:\Users\SumitJain\Documents\upgd\recommendation_system\data\consumer_transanctions.csv')

    return plat_articles,user_behaviour_df

plat_articles,user_behaviour_df = reading_content_data()

plat_articles.shape
plat_articles.head(2)
```

#### Merging consumer and platform content metadata to come up with a common dataset containing all attributes.


```python
merged_inner = pd.merge(left=plat_articles, right=user_behaviour_df, left_on='item_id', right_on='item_id')

# size of the output data

merged_inner.shape
merged_inner.head(3)
```

#### getting interaction_weightages from configs and applying on each interaction type to get type weightages


```python
user_behaviour_df['type_weightage'] = user_behaviour_df['interaction_type'].apply(lambda x: interaction_weightages[x])
```

```python
user_behaviour_df
```

#### getting content consumed by each users as well as getting users with more than 5 items to understand if there are enough 
#### users with good enough engagement as evident below


```python
consumer_behaviour_count_df = user_behaviour_df.groupby(['consumer_id', 'item_id']).size().groupby('consumer_id').size()
consumer_behaviour_count_df


print('no. of users: %d' % len(consumer_behaviour_count_df))
users_with_enough_user_behaviour_df = consumer_behaviour_count_df[consumer_behaviour_count_df >= 5].reset_index()[['consumer_id']]
print('no. of users with atleast 5 consumptions: %d' % len(users_with_enough_user_behaviour_df))
```

#### getting only those interactions which has >5 items consumed by users for better model performance in future


```python
print('no. of interactions: %d' % len(user_behaviour_df))
user_behaviour_from_selected_users_df = user_behaviour_df.merge(users_with_enough_user_behaviour_df, how = 'right',left_on = 'consumer_id',
right_on = 'consumer_id')
print('no. of interactions from users with at least 5 engagement: %d' % len(user_behaviour_from_selected_users_df))
```

```python
user_behaviour_full_df = user_behaviour_from_selected_users_df[['consumer_id', 'item_id','type_weightage']]
user_behaviour_full_df.to_csv(r'C:\Users\SumitJain\Documents\upgd\recommendation_system\data\data_rating.csv')
```

#### normalizing users event types


```python
user_behaviour_full_df = user_behaviour_from_selected_users_df.groupby(['consumer_id', 'item_id'])['type_weightage'].sum() \
.apply(lambda x: math.log(1+x, 2)).reset_index() # lambda x: math.log(1+x, 2)
print('no. of unique user/item interactions: %d' % len(user_behaviour_full_df))
user_behaviour_full_df.head(10)
# user_behaviour_full_df.to_csv("data.csv")
```

#### Get the user's data and merge in the content info for modelling usage


```python
def get_users_content_data(person_id, user_behaviour_df):
    interacted_items = user_behaviour_df.loc[person_id]['item_id']
    return set(interacted_items if type(interacted_items) == pd.Series else [interacted_items])
```

#### spliting to train test data


```python
def train_test_user_behaviour():
    user_behaviour_train_df, user_behaviour_test_df = train_test_split(user_behaviour_full_df,test_size=0.15,random_state=22)
    # stratify=user_behaviour_full_df['consumer_id'], 
    print('engagement in train set: %d' % len(user_behaviour_train_df))
    print('engagement in on test set: %d' % len(user_behaviour_test_df))
    
    return user_behaviour_train_df, user_behaviour_test_df

#Indexing by consumer_id to speed up the searches during evaluation
user_behaviour_train_df, user_behaviour_test_df = train_test_user_behaviour()
print(user_behaviour_train_df)
```

#### getting_user_behaviour_indexed for modelling usage


```python
def getting_user_behaviour_indexed():
    
    user_behaviour_full_indexed_df = user_behaviour_full_df.set_index('consumer_id')
    user_behaviour_train_indexed_df = user_behaviour_train_df.set_index('consumer_id')
    user_behaviour_test_indexed_df = user_behaviour_test_df.set_index('consumer_id')
#     print(user_behaviour_train_indexed_df)
    return user_behaviour_full_indexed_df,user_behaviour_train_indexed_df,user_behaviour_test_indexed_df

getting_user_behaviour_indexed()
```
