### Collaborative Filtering on User Engagement Data


#### loading required dependencies

```python
import os
import sys
import import_ipynb
import numpy as np
import pandas as pd
sys.path.append(os.path.abspath(os.path.join('__file__','../../')))
from modules import evaluations 
from modules import data_processing
from scipy.sparse import csr_matrix
from scipy.sparse.linalg import svds
from config.configs import NUMBER_OF_FACTORS_MF,MODEL_COLLAB
```

#### loading required data from data processing code and creating a pivot table

```python
# generate a sparse pivot table with users in rows and items in columns

user_behaviour_train_df = data_processing.train_test_user_behaviour()[0]
plat_articles = data_processing.reading_content_data()[0]
users_items_pivot_matrix_df = user_behaviour_train_df.pivot(index='consumer_id', columns='item_id',
                                                          values='type_weightage').fillna(0)

users_items_pivot_matrix_df.head(10)
```

```python
users_items_pivot_matrix = users_items_pivot_matrix_df.as_matrix()
users_items_pivot_matrix[:1]
```

#### creating a csr matrix 

```python
users_ids = list(users_items_pivot_matrix_df.index)
print("users_ids:\n",users_ids[:10])

users_items_pivot_sparse_matrix = csr_matrix(users_items_pivot_matrix)
# print("\nusers_items_pivot_sparse_matrix: \n",users_items_pivot_sparse_matrix)
```

#### Performing matrix factorization on the original user item matrix


```python
#user_matrix, sigma, item_matrix = svds(users_items_pivot_matrix, k = NUMBER_OF_FACTORS_MF)
user_matrix, sigma, item_matrix = svds(users_items_pivot_sparse_matrix, k = NUMBER_OF_FACTORS_MF)
```

```python
user_matrix.shape
item_matrix.shape
```

```python
sigma = np.diag(sigma)
sigma.shape
sigma
```

#### dot product between user sigma item matrices would result an array to calculate ratings

```python
all_user_predicted_ratings = np.dot(np.dot(user_matrix, sigma), item_matrix) 
all_user_predicted_ratings
```

```python
all_user_predicted_ratings_norm = (all_user_predicted_ratings - all_user_predicted_ratings.min()) / (all_user_predicted_ratings.max() - all_user_predicted_ratings.min())
```

#### Converting the reconstructed matrix back to a Pandas dataframe

```python
cf_preds_df = pd.DataFrame(all_user_predicted_ratings_norm, columns = users_items_pivot_matrix_df.columns, index=users_ids).transpose()
cf_preds_df.head(10)
# len(cf_preds_df.columns)
```

```python
def get_ratings_df():
    
    #Converting the reconstructed matrix back to a Pandas dataframe
    
    cf_preds_df = pd.DataFrame(all_user_predicted_ratings_norm, columns = users_items_pivot_matrix_df.columns, index=users_ids).transpose()
    cf_preds_df.head(1)
    return cf_preds_df
    
```

#### creating the final cf model, ratings and recommendation

```python
class CollabFiltering:
    
    
    def __init__(self, svd_matrix_df, items_df=None):
        self.svd_matrix_df = svd_matrix_df
        self.items_df = items_df

    def get_item_recommendations(self, user_id, items_to_ignore=[], topn=10, verbose=False):
        # Get and sort the user's predictions
        sorted_user_predictions = self.svd_matrix_df[user_id].sort_values(ascending=False) \
                                    .reset_index().rename(columns={user_id: 'sim_score'})

        # Recommend the highest predicted rating news that the user hasn't seen yet.
        recommendations_df = sorted_user_predictions[~sorted_user_predictions['item_id'].isin(items_to_ignore)] \
                               .sort_values('sim_score', ascending = False) \
                               .head(topn)

        if verbose:
            if self.items_df is None:
                raise Exception('"items_df" is required in verbose mode')

            recommendations_df = recommendations_df.merge(self.items_df, how = 'left', 
                                                          left_on = 'item_id', 
                                                          right_on = 'item_id')[['sim_score', 'item_id', 'title', 'item_url', 'language']]


        return recommendations_df
    
svd_preds_df=get_ratings_df()    
recommender_cf = CollabFiltering(svd_preds_df, plat_articles)
```

#### calling evaluation code for evaluating SVD Matrix Factorization model performance

```python
collab_overall_metrics, collab_eval_results_df = evaluations.model_evaluator.evaluate_model(recommender_cf,MODEL_COLLAB)
print('overall metrics:\n', collab_overall_metrics)
collab_eval_results_df.head(10)
```
