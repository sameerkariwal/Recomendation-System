#### getting collaborative and content based models together to generate recommendations through a hybrid model

```python
class HybridRecommender:
    
    
    def __init__(self, content_model, collab_model, items_df, content_model_weight=1.0, collab_model_weight=1.0):
        
        self.content_model = content_model
        self.collab_model = collab_model
        self.content_model_weight = content_model_weight
        self.collab_model_weight = collab_model_weight
        self.items_df = items_df
        
    def get_item_recommendations(self, user_id, items_to_ignore=[], topn=10, verbose=False):
        #Getting the top-1000 Content-based filtering recommendations
        content_recom_df = self.content_model.get_item_recommendations(user_id, items_to_ignore=items_to_ignore, verbose=verbose,
                                                           topn=1000).rename(columns={'sim_score': 'sim_score_content'})
        
        #Getting the top-1000 Collaborative filtering recommendations
        collab_recom_df = self.collab_model.get_item_recommendations(user_id, items_to_ignore=items_to_ignore, verbose=verbose, 
                                                           topn=1000).rename(columns={'sim_score': 'sim_score_collab'})
        
        #Combining the results by item_id
        recom_df = content_recom_df.merge(collab_recom_df,
                                   how = 'outer', 
                                   left_on = 'item_id', 
                                   right_on = 'item_id').fillna(0.0)
        
        #Computing a hybrid recommendation score based on collab and content scores
        #recom_df['hybrid_rec_relevance_score'] = recom_df['sim_score_content'] * recom_df['sim_score_collab'] 
        recom_df['hybrid_rec_relevance_score'] = (recom_df['sim_score_content'] * self.content_model_weight) \
                                     + (recom_df['sim_score_collab'] * self.collab_model_weight)
        
        #Sorting recommendations by hybrid score
        recommendations_df = recom_df.sort_values('hybrid_rec_relevance_score', ascending=False).head(topn)

        if verbose:
            if self.items_df is None:
                raise Exception('"items_df" is required in verbose mode')

            recommendations_df = recommendations_df.merge(self.items_df, how = 'left', 
                                                          left_on = 'item_id', 
                                                          right_on = 'item_id')[['hybrid_rec_relevance_score', 'item_id', 'title', 'item_url', 'language']]


        return recommendations_df

    
plat_articles = data_processing.reading_content_data()[0]
collab_preds_df=collaborative_filtering_model.get_ratings_df()
content_based_recommender_model = content_based_filtering_model.ContentFiltering(plat_articles)
recommender_collab = collaborative_filtering_model.CollabFiltering(collab_preds_df, plat_articles)
hybrid_recommender_model = HybridRecommender(content_based_recommender_model, recommender_collab, plat_articles,
                                             content_model_weight=1.0, collab_model_weight=100.0)
```

#### evaluating final hybrid model performance

```python
hybrid_overall_metrics, hybrid_eval_results_df = evaluations.model_evaluator.evaluate_model(hybrid_recommender_model,MODEL_HYBRID)
print('overall metrics:\n', hybrid_overall_metrics)
hybrid_eval_results_df.head(10)
```
